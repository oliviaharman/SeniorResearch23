```python
# Open the file
import fitz
file = "Largen & Spawls 2010 part VI snakes.pdf"
pdf = fitz.open(file)
```

```python
# Make sure it imported correctly
print(pdf.page_count)
```

```python
counter = 1
for i in range(len(pdf)):
    page = pdf[i]
    images = page.get_images()
    for image in images:
        base_img = pdf.extract_image(image[0])
        image_data =base_img["image"]
        img = PIL.Image.open(io.BytesIO(image_data))
        extension = base_img["ext"]
        img.save(open(f"image{counter}.{extension}", "wb"))
        counter += 1
```

```python
# Change the variables to rect
from pdfminer.high_level import extract_pages
from pdfminer.layout import LTTextContainer, LTChar
for page_layout in extract_pages("Largen & Spawls 2010 part VI snakes.pdf"):
    for element in page_layout:
        if isinstance(element, LTTextContainer):
            for text_line in element:
                for character in text_line:
                    if isinstance(character, LTChar):
                        print(character.fontname)
                        print(character.size)
                        
# It worked!!!!! - wait was it this one or the other one?
# delete pictures and restart the kernel to see - at first it was only a select few pages
```
